```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ploting basic plots
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
x = np.array([0,5])
y = np.array([0,25])

plt.plot(x, y)
plt.show()
```

```python
x = np.array([0,5,10,12,18])
y = np.array([0,25,28,30,32])

## Adding Style 
plt.plot(x, y, 'ro')
plt.show()
```

```python
## Adding Style 
plt.plot(x, y, color='red',marker='*')
plt.show()
```

```python
## Adding Style
## Adding label
plt.plot(x, y, color='red',marker='*')
plt.ylabel('Quantity')
plt.xlabel('Quality')
plt.show()
```

```python
## Adding Style
## Adding Title
plt.plot(x, y, color='red',marker='*')
plt.ylabel('Quantity')
plt.xlabel('Quality')
plt.title('Quantity vs Quality plot')
plt.show()
```

```python
## Adding Style
## Adding legend
### where FIG means figure:things outside the table and ax means axis: things within or insode the table.
fig,ax = plt.subplots()
ax.plot(x,y*2, label='first')
ax.plot(y,x*2, label= 'second')
ax.plot(y,x, label= 'third')
ax.plot(x,y, label= 'fourth')
ax.legend()
```

```python
## Adding Style
## Adding Grid
fig,ax = plt.subplots()
ax.plot(x,y*2, label='first')
ax.plot(y,x*2, label= 'second')
ax.plot(y,x, label= 'third')
ax.plot(x,y, label= 'fourth')
ax.legend()
ax.grid(True)
plt.show()
```

```python
df = pd.read_csv("Datasets/Diabetes Analysis/diabetes.csv")
df.head()
```

## Line Plot

```python
#Line Plot

#Creating a figure and axis object
fig, ax = plt.subplots()

##Displaying Data

ax.plot(df['Pregnancies'], df['Glucose'],marker='*')

##Adding labels
ax.set_xlabel("Pregnancies")
ax.set_ylabel("Glucose")
ax.set_title("Pregnancy vs Glucose")

## Display
plt.show()
```

```python
#Scatter plot
#Creating a figure and axis object
fig, ax = plt.subplots()

##Displaying Data

ax.scatter(df['Pregnancies'], df['Glucose'])

##Adding labels
ax.set_xlabel("Pregnancies")
ax.set_ylabel("Glucose")
ax.set_title("Pregnancy vs Glucose")

## Display
plt.show()
```

```python
#Histogran plot
#Creating a figure and axis object
fig, ax = plt.subplots()

##Displaying Data

ax.hist (df['Outcome'])
         
##Adding labels
ax.set_xlabel("Outcome")

ax.set_title("Diabetes Outcome")

ax.set_xticks([0,1])

         
## Display
plt.show()
```

```python
plt.bar(df['Pregnancies'],df['BMI'], color= 'r',label='pregnancy')
plt.xlabel('Pregnancies')
plt.ylabel('BMI')
plt.title('Pregnancy vs BMI')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.bar(df['Pregnancies'],df['BMI'], color= 'r',label='pregnancy')
plt.bar(df['Age'],df['BMI'], color= 'green',label='pregnancy')
plt.bar(df['Glucose'],df['BMI'], color= 'yellow',label='pregnancy')

# plt.xlabel('Pregnancies')
# plt.ylabel('BMI')
# plt.title('Pregnancy vs BMI')



plt.legend()
plt.grid()
plt.show()
```

```python
#Stacked Bar Chart

plt.bar(df['Age'],df['BloodPressure'], color= 'r',label='BloodPressure')
plt.bar(df['Age'],df['SkinThickness'], color= 'green',label='SkinThickness')
plt.bar(df['Age'],df['BMI'], color= 'yellow',label='BMI')

plt.ylabel("Age")
plt.title("STACKED BAR CHART")

plt.legend()

plt.grid()

plt.show()
```

```python
# Pie Chart
a = np.array([2,6,12,10,4,5])
labels = ['Rivers','Abia','Imo','Enugu','Benue','Nasarawa']
colors = ['Black','Lightgray','Gray','Darkgray','Silver','Dimgray']
explode = (0,0,0,0,0.1,0)



fig,ax = plt.subplots()
ax.pie(a,labels=labels, colors=colors,shadow=True, explode=explode)
plt.show()
```

```python
# Pie Chart
a = np.array([2,6,12,10,4,5])
labels = ['Rivers','Abia','Imo','Enugu','Benue','Nasarawa']
colors = ['Black','Lightgray','Gray','Darkgray','Silver','Dimgray']
explode = (0,0,0,0,0,0.4)
hatch = ['','**','','','','']


fig,ax = plt.subplots()
ax.pie(a,labels=labels,
       colors=colors,
       shadow=True,
       explode=explode,
       hatch=hatch,
       autopct='%1.2f%%') ## automatic percentage
plt.show()
```
